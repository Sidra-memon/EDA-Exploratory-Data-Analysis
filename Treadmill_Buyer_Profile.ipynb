```python
import seaborn as sns
#the count is a simple frequency measure that tells you how many data points are in each bin.
#This allows you to quickly understand the distribution of the data across different intervals
# Features to plot
features = ['Age', 'Education', 'Usage', 'Fitness', 'Income', 'Miles']

# Create a figure with 2 rows and 3 columns
fig, axes = plt.subplots(2, 3, figsize=(18, 10))  # 2 rows, 3 columns

# Flatten axes for easier indexing
axes = axes.flatten()

# Loop through features and create a histogram for each
for i, feature in enumerate(features):
    sns.histplot(df[feature], kde=False, bins=10, ax=axes[i])
    axes[i].set_title(f"Distribution of {feature}")
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel("Frequency")


# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```

Outlier Detection:

*   Check for the outliers by using the IQR method.




```python
percentile25 = df['Education'].quantile(0.25)

percentile75 = df['Education'].quantile(0.75)

iqr = percentile75 - percentile25

upper_limit = percentile75 + 1.5 * iqr

lower_limit = percentile25 - 1.5 * iqr

upper_limit, lower_limit

new_df = df[(df['Education'] > upper_limit) | (df['Education'] < lower_limit)]
num_outliers = len(new_df)
num_outliers
```

# Conditional Probabilities:

What percent of customers have purchased KP281, KP481, or KP781?

---



```python
# Total number of customers
total_customers = len(df)

# Filter for the specific products: KP281, KP481, KP781
filtered_df = df[df['Product'].isin(['KP281', 'KP481', 'KP781'])]

# Group by Product and count occurrences
grouped_data = filtered_df.groupby('Product').size().reset_index(name='Count')

# Add a percentage column
grouped_data['Percentage'] = (grouped_data['Count'] / total_customers) * 100

# Display the grouped data
print(grouped_data)
```

Observation\
**KP281** is the dominant product, preferred by the majority of customers.
**KP481** and **KP781** follow, but with significantly smaller shares in comparison to **KP281**.
The dataset shows a clear preference for **KP281**, which might be useful for marketing and sales strategies targeting the more popular product.

Create frequency tables and calculate the percentage as follows\


*   Product - Gender

1. Percentage of a Male customer purchasing a treadmill
2. Percentage of a Female customer purchasing KP781 treadmill
1.  Probability of a customer being a Female given that Product is KP281





```python
# prompt: Percentage of a Male and female customer purchasing a treadmill

# Calculate the percentage of male customers purchasing a treadmill
male_customers = df[df['Gender'] == 'Male']
total_customers = len(df)
male_percentage = (len(male_customers) / total_customers) * 100
print(f"Percentage of Male Customers: {male_percentage:.2f}%")

# Calculate the percentage of female customers purchasing a KP781 treadmill
female_kp781_customers = df[(df['Gender'] == 'Female') & (df['Product'] == 'KP781')]
female_percentage = (len(female_kp781_customers) / len(df[df['Gender'] == 'Female'])) * 100
print(f"Percentage of Female Customers purchasing KP781: {female_percentage:.2f}%")

# Calculate the probability of a customer being female given they purchased a KP281 treadmill
kp281_customers = df[df['Product'] == 'KP281']
female_kp281_customers = kp281_customers[kp281_customers['Gender'] == 'Female']
probability = len(female_kp281_customers) / len(kp281_customers)
print(f"Probability of a customer being Female given they purchased KP281: {probability:.2f}")
```

# Actionable Insights & Recommendations

 From the analysis, here are some actionable insights and recommendations based on the percentages calculated:

**Treadmill Popularity:**

KP281 is the most popular treadmill among customers, with 44.44% of all customers purchasing this product. This suggests that the marketing focus could be aligned with promoting KP281 more prominently.
KP481 follows with 33.33%, while KP781 has a smaller market share at 22.22%. Considering these trends, strategies should include targeting customers who prefer products with lower market share (like KP781) to boost sales.\

**Customer Demographics:**

Gender: Approximately 57.78% of customers are male, with 42.22% being female. Understanding this gender distribution can help in tailoring marketing messages and campaigns.
Age: The majority of customers fall in the 20s and 30s, with 87.78% of customers falling into this age group. This suggests that marketing efforts should be focused on younger adults who are more likely to purchase a treadmill.
Marital Status: 59.44% of customers are partnered, indicating that treadmill purchases are more common among individuals in relationships. This might suggest that marketing strategies should also consider the concept of family-oriented health and fitness.\

**Income:**

Low-Income Customers (below $40,000): Only 17.78% of low-income customers purchased a treadmill. On the other hand, high-income customers (above $40,000) have a higher propensity for purchasing specific models, especially KP781, with 27.03% of high-income customers purchasing it.
A large percentage of high-income customers buying KP781 suggests that this treadmill might be positioned as a premium product. This could guide product positioning strategies.

**Fitness Level:**

17.22% of customers have a fitness level of 5 (highest fitness level). Among these, 93.55% are purchasing KP781, showing a strong correlation between high fitness levels and the preference for the KP781 treadmill.
Tailoring marketing to fitness enthusiasts with high fitness levels could increase sales for the KP781.

**Recommendations:**

**Product Promotion:** Focus more on promoting the KP281 while also creating campaigns targeted towards high-income customers for KP781.

**Age & Gender Targeting:** Cater to the younger demographic (20s-30s) with gender-specific promotions, given the higher engagement of males and females in the treadmill market.

**Income-Based Strategies:** For low-income customers, consider offering discounts, financing options, or emphasizing the affordability of other treadmill models.

**Fitness-Centric Marketing:** Market KP781 as the go-to treadmill for serious fitness enthusiasts, especially those with higher fitness levels.

**Marital Status Focus:** Given the high percentage of partnered customers, consider promoting family or couple-themed fitness packages.
